#  **1. Executive Summary**




This study utilizes crash reporting datasets on Drivers and Non-Motorists from Montgomery County, Maryland to provide insight into crash occurrence and pedestrian involvement across the county. The goal of this study is to help the public and the Montgomery County Government better understand the factors that contribute to crash incidents and to support data-driven recommendations to reduce and prevent future crashes. Through cross-analysis of temporal trends, environmental factors, infrastructural conditions, and crash responsibility, findings suggest that no single factor alone leads to crash reports. Rather, it is the overlap of multiple conditions that increases the likelihood of an incident. These findings led to the development of the Road Risk Severity Index (RRSI), which assigns weighted risk scores to specific roads based on crash frequency, injury severity, and time-of-day. # **2. Introduction**

### **Problem Definition**

This project analyzes traffic crash patterns in Montgomery County, Maryland by integrating driver and non-motorist datasets to understand the factors that influence crash likelihood and severity. It examines how driver behavior, pedestrian actions, and environmental conditions such as location, time of day, weather, and road characteristics shape crash outcomes.

The analysis focuses on:

*   Crash timing and trends over the years, including differences across crash types (property damage, injury, and fatal).
*   Environmental and roadway conditions weather, lighting, surface condition, and traffic controls, and their impact on crash frequency and severity.

*   Driver and non-motorist behaviors, identifying who is more often at fault and how responsibility varies between vehicle involved and pedestrian-related crashes.

Who cares about this problem, and why?

* County transportation agencies (MCDOT, SHA): Need data-driven evidence to identify high-risk road segments and prioritize safety investments.

* Law enforcement: Uses crash timing, risky behaviors, and substance involvement patterns to guide targeted enforcement during peak crash hours.

* Urban planners & policymakers: Depend on insights about weather, lighting, roadway conditions, and traffic controls to design safer streets and pedestrian infrastructure.

* Public health & community groups: Focus on protecting pedestrians, cyclists, and other vulnerable road users who are disproportionately impacted by unsafe conditions.

* Insurance companies & fleet operators: Use crash trends and collision types to assess risk exposure and improve safety protocols.

```python
_sql = """
CREATE TEMP FUNCTION to_int64(s STRING) AS (
  SAFE_CAST(NULLIF(REGEXP_REPLACE(s, r'[^0-9-]', ''), '') AS INT64)
);
CREATE TEMP FUNCTION to_bool(s STRING) AS (
  CASE
    WHEN REGEXP_CONTAINS(LOWER(s), r'^(y|yes|true|1)$')  THEN TRUE
    WHEN REGEXP_CONTAINS(LOWER(s), r'^(n|no|false|0)$') THEN FALSE
    ELSE NULL
  END
);
CREATE TEMP FUNCTION to_datetime(s STRING) AS (
  CASE
    WHEN REGEXP_CONTAINS(s, r'^\d{{1,2}}/\d{{1,2}}/\d{{2}}\s+\d{{1,2}}:\d{{2}}(:\d{{2}})?$')
      THEN PARSE_DATETIME('%m/%d/%y %H:%M', s)
    WHEN REGEXP_CONTAINS(s, r'^\d{{1,2}}/\d{{1,2}}/\d{{4}}\s+\d{{1,2}}:\d{{2}}(:\d{{2}})?$')
      THEN PARSE_DATETIME('%m/%d/%Y %H:%M', s)
    WHEN REGEXP_CONTAINS(s, r'^\d{{4}}-\d{{2}}-\d{{2}}\s+\d{{2}}:\d{{2}}:\d{{2}}')
      THEN PARSE_DATETIME('%Y-%m-%d %H:%M:%S', s)
    ELSE NULL
  END
);

CREATE OR REPLACE TABLE `ba775-fall25-b04.crash_reporting_dataset.drivers_data_clean` AS
WITH geo_parsed AS (
  SELECT
    t.*,
    SAFE_CAST(`Latitude`  AS FLOAT64) AS lat_num,
    SAFE_CAST(`Longitude` AS FLOAT64) AS lon_num,

    SAFE_CAST(REGEXP_EXTRACT(`Location`, r'\(\s*([+-]?\d+(?:\.\d+)?)') AS FLOAT64)   AS loc_lat,
    SAFE_CAST(REGEXP_EXTRACT(`Location`, r',\s*([+-]?\d+(?:\.\d+)?)\s*\)') AS FLOAT64) AS loc_lon
  FROM `ba775-fall25-b04.crash_reporting_dataset.drivers_data` AS t
)
SELECT
  *
  REPLACE (
    to_datetime(`Crash Date Time`) AS `Crash Date Time`,
    to_int64(`Speed Limit`)        AS `Speed Limit`,
    to_bool(`Driverless Vehicle`)  AS `Driverless Vehicle`,
    to_bool(`Parked Vehicle`)      AS `Parked Vehicle`,
    COALESCE(lat_num, loc_lat)     AS `Latitude`,
    COALESCE(lon_num, loc_lon)     AS `Longitude`
  ),
  CASE
    WHEN COALESCE(lat_num, loc_lat) IS NOT NULL
     AND COALESCE(lon_num, loc_lon) IS NOT NULL
    THEN ST_GEOGPOINT(COALESCE(lon_num, loc_lon), COALESCE(lat_num, loc_lon))
    ELSE NULL
  END AS location_geog
FROM geo_parsed;
""" # end _sql
from google.colab.sql import bigquery as _bqsqlcell
df = _bqsqlcell.run(_sql)
df
```

```python
# sql_engine: bigquery
# output_variable: df
# start _sql
_sql = r"""
CREATE OR REPLACE TABLE `ba775-fall25-b04.crash_reporting_dataset.drivers_data_clean2` AS
SELECT
  -- 1. Keep all original columns EXCEPT the old 'weather' one
  * EXCEPT(weather),

  -- 2. Clean the 'weather' column
  CASE
    -- Standardize Clear (handling spaces)
    WHEN UPPER(TRIM(weather)) = 'CLEAR' THEN 'Clear'

    -- Standardize Cloudy
    WHEN UPPER(TRIM(weather)) = 'CLOUDY' THEN 'Cloudy'

    -- Standardize Rain
    WHEN UPPER(TRIM(weather)) IN ('RAIN', 'RAINING') THEN 'Rain'

    -- Standardize Fog
    WHEN UPPER(TRIM(weather)) IN ('FOG, SMOG, SMOKE', 'FOGGY') THEN 'Fog, Smog, Smoke'

    -- Standardize Snow
    WHEN UPPER(TRIM(weather)) = 'SNOW' THEN 'Snow'

    -- Standardize Sleet/Hail
    WHEN UPPER(TRIM(weather)) IN ('SLEET OR HAIL', 'SLEET') THEN 'Sleet or Hail'

    -- Standardize Blowing Snow
    WHEN UPPER(TRIM(weather)) = 'BLOWING SNOW' THEN 'Blowing Snow'

    -- Standardize Wind
    WHEN UPPER(TRIM(weather)) IN ('SEVERE CROSSWINDS', 'SEVERE WINDS') THEN 'Severe Crosswinds'

    -- Standardize Wintry Mix
    WHEN UPPER(TRIM(weather)) = 'WINTRY MIX' THEN 'Wintry Mix'

    -- Standardize Freezing Rain
    WHEN UPPER(TRIM(weather)) = 'FREEZING RAIN OR FREEZING DRIZZLE' THEN 'Freezing Rain or Drizzle'

    -- Standardize Blowing Sand (partial match)
    WHEN UPPER(TRIM(weather)) LIKE '%BLOWING SAND%' THEN 'Blowing Sand, Soil, Dirt'

    -- Filter out UUIDs (Codes like 07735BE9...)
    WHEN REGEXP_CONTAINS(TRIM(weather), r'^[0-9A-Fa-f]{{8}}-[0-9A-Fa-f]{{4}}-[0-9A-Fa-f]{{4}}-[0-9A-Fa-f]{{4}}-[0-9A-Fa-f]{{12}}$') THEN NULL

    -- Filter out non-weather terms
    WHEN UPPER(TRIM(weather)) IN (
      'NO APPARENT INJURY', 'UNKNOWN', 'OTHER', 'YES', 'NO',
      'SUSPECTED MINOR INJURY', 'POSSIBLE INJURY', 'SUSPECTED SERIOUS INJURY', 'N/A'
    ) THEN NULL

    ELSE NULL
  END AS weather
FROM
  `ba775-fall25-b04.crash_reporting_dataset.drivers_data_clean`;
""" # end _sql
from google.colab.sql import bigquery as _bqsqlcell
df = _bqsqlcell.run(_sql)
df
```

### **Merging**


```python
# sql_engine: bigquery
# output_variable: df
# start _sql
_sql = """
CREATE OR REPLACE TABLE `ba775-fall25-b04.crash_reporting_dataset.merged_data` AS
SELECT
  COALESCE(d.`Report Number`, n.`Report Number`) AS report_number,

  COALESCE(CAST(d.`Crash Date Time` AS DATETIME), CAST(n.`Crash Date Time` AS DATETIME)) AS crash_datetime,
  COALESCE(d.`ACRS Report Type`, n.`ACRS Report Type`) AS acrs_report_type,
  COALESCE(d.`Route Type`, n.`Route Type`) AS route_type,
  COALESCE(d.`Road Name`, n.`Road Name`) AS road_name,
  COALESCE(d.`Cross-Street Name`, n.`Cross-Street Name`) AS cross_street_name,
  COALESCE(d.`Related Non-Motorist`, n.`Related Non-Motorist`) AS related_non_motorist,
  COALESCE(d.`Collision Type`, n.`Collision Type`) AS collision_type,
  COALESCE(d.`Weather`, n.`Weather`) AS weather,
  COALESCE(d.`Surface Condition`, n.`Surface Condition`) AS surface_condition,
  COALESCE(d.`Light`, n.`Light`) AS light,
  COALESCE(d.`Traffic Control`, n.`Traffic Control`) AS traffic_control,
  COALESCE(d.`Driver Substance Abuse`, n.`Driver Substance Abuse`) AS driver_substance_abuse,
  COALESCE(d.`Non-Motorist Substance Abuse`, n.`Non-Motorist Substance Abuse`) AS non_motorist_substance_abuse,
  COALESCE(d.Latitude, n.Latitude) AS latitude,
  COALESCE(d.Longitude, n.Longitude) AS longitude,
  COALESCE(d.Location, n.Location) AS location,

  d.`Local Case Number`                 AS driver_local_case_number,
  d.`Agency Name`                       AS driver_agency_name,
  d.`Driver At Fault`                   AS driver_at_fault,
  d.`Injury Severity`                   AS driver_injury_severity,
  d.`Circumstance`                      AS driver_circumstance,
  d.`Driver Distracted By`              AS driver_distracted_by,
  d.`Drivers License State`             AS driver_license_state,
  d.`Vehicle ID`                        AS driver_vehicle_id,
  d.`Vehicle Damage Extent`             AS driver_vehicle_damage_extent,
  d.`Vehicle First Impact Location`     AS driver_vehicle_first_impact_location,
  d.`Vehicle Body Type`                 AS driver_vehicle_body_type,
  d.`Vehicle Movement`                  AS driver_vehicle_movement,
  d.`Vehicle Going Dir`                 AS driver_vehicle_going_dir,
  d.`Speed Limit`                       AS driver_speed_limit,
  d.`Driverless Vehicle`                AS driver_driverless_vehicle,
  d.`Parked Vehicle`                    AS driver_parked_vehicle,
  d.`Vehicle Year`                      AS driver_vehicle_year,
  d.`Vehicle Make`                      AS driver_vehicle_make,
  d.`Vehicle Model`                     AS driver_vehicle_model,

  n.`Pedestrian Type`                   AS nm_pedestrian_type,
  n.`Pedestrian Movement`               AS nm_pedestrian_movement,
  n.`Pedestrian Actions`                AS nm_pedestrian_actions,
  n.`Pedestrian Location`               AS nm_pedestrian_location,
  n.`At Fault`                          AS nm_at_fault,
  n.`Safety Equipment`                  AS nm_safety_equipment,
  n.`Injury Severity`                   AS nm_injury_severity

FROM `ba775-fall25-b04.crash_reporting_dataset.drivers_data_clean` AS d
FULL OUTER JOIN `ba775-fall25-b04.crash_reporting_dataset.non_motorists_clean ` AS n
  ON d.`Report Number` = n.`Report Number`;
""" # end _sql
from google.colab.sql import bigquery as _bqsqlcell
df = _bqsqlcell.run(_sql)
df
```

This merge code performs a FULL OUTER JOIN on the drivers_data_clean and non_motorists_clean tables using the Report Number column. This is because we wanted to combine all records from both tables, even if there isn't a match in the other table. For common fields, it uses COALESCE to pick the value from the drivers' table if it exists, otherwise it uses the value from the non-motorists' table. It also renames many columns to make it clear whether they originated from the driver's data or the non-motorist's data. This effectively creates a unified crash-level table with separate, clearly identified details for drivers and non-motorists.

**OVERALL SUMMARY:**

We cleaned and standardized both datasets by safely converting messy string inputs into proper data types, fixing inconsistent text formats, and replacing uninformative categories with NULL to reduce noise. Key fields such as dates, numeric values, boolean indicators, and geographic coordinates were parsed and reconstructed to ensure accuracy and consistency across records. Finally, we merged the driver and non-motorist tables using a FULL OUTER JOIN with clear column labeling, creating a unified crash-level dataset ready for reliable analysis.

### **4.3.3 How does the posted speed limit relate to the distribution of injury severity in reported crashes**

i) Understanding how injury outcomes shift as roadway speed limits increase

```python
# sql_engine: bigquery
# output_variable: df
# start _sql
_sql = """
WITH cleaned AS (
  SELECT
    driver_speed_limit,
    CASE
      WHEN driver_injury_severity IS NULL THEN 'Unknown'

      WHEN LOWER(driver_injury_severity) LIKE 'no apparent injury%'
        THEN 'No Apparent Injury'

      WHEN LOWER(driver_injury_severity) LIKE 'possible injury%'
        THEN 'Possible Injury'

      WHEN LOWER(driver_injury_severity) LIKE 'suspected minor injury%'
        THEN 'Suspected Minor Injury'

      WHEN LOWER(driver_injury_severity) LIKE 'suspected serious injury%'
        THEN 'Suspected Serious Injury'

      WHEN LOWER(driver_injury_severity) LIKE 'fatal injury%'
        THEN 'Fatal Injury'

      ELSE driver_injury_severity
    END AS injury_severity_clean
  FROM `ba775-fall25-b04.crash_reporting_dataset.merged_data`
  WHERE driver_speed_limit BETWEEN 5 AND 75
    AND driver_injury_severity IS NOT NULL
)

SELECT
  driver_speed_limit,
  injury_severity_clean AS driver_injury_severity,
  COUNT(*) AS crash_count,
  ROUND(
    100 * SAFE_DIVIDE(
      COUNT(*),
      SUM(COUNT(*)) OVER (PARTITION BY driver_speed_limit)
    ),
    1
  ) AS weight_in_speed_limit
FROM cleaned
GROUP BY driver_speed_limit, injury_severity_clean
ORDER BY driver_speed_limit, injury_severity_clean;

""" # end _sql
from google.colab.sql import bigquery as _bqsqlcell
df = _bqsqlcell.run(_sql)
df
```

**Answer:** Across all posted speed limits, most crashes result in no apparent injury, but as speeds increase, the distribution gradually shifts toward more harmful outcomes. Minor and possible injuries become slightly more common in mid-speed environments, and although severe injuries remain rare overall, they appear more frequently where vehicles travel faster. This indicates that higher operating speeds subtly but consistently elevate the likelihood of more serious injury outcomes.

ii) Comparing severe injury likelihood across speed environments

```python
# sql_engine: bigquery
# output_variable: df
# start _sql
_sql = """
WITH speed_binned AS (
  SELECT
    CASE
      WHEN driver_speed_limit BETWEEN 5 AND 25 THEN '05–25 mph'
      WHEN driver_speed_limit BETWEEN 30 AND 35 THEN '30–35 mph'
      WHEN driver_speed_limit BETWEEN 40 AND 45 THEN '40–45 mph'
      WHEN driver_speed_limit BETWEEN 50 AND 55 THEN '50–55 mph'
      WHEN driver_speed_limit BETWEEN 60 AND 75 THEN '60–75 mph'
      ELSE 'Other'
    END AS speed_bin,
    driver_injury_severity
  FROM `ba775-fall25-b04.crash_reporting_dataset.merged_data`
  WHERE driver_speed_limit BETWEEN 5 AND 75
    AND driver_injury_severity IS NOT NULL
)
SELECT
  speed_bin,
  COUNT(*) AS total_crashes,
  SUM(CASE
        WHEN driver_injury_severity IN ('Suspected Serious Injury', 'Fatal Injury')
        THEN 1 ELSE 0
    END
  ) AS severe_crashes,
  ROUND(
    100 * SAFE_DIVIDE(
      SUM(
        CASE
          WHEN driver_injury_severity IN ('Suspected Serious Injury', 'Fatal Injury')
          THEN 1 ELSE 0
        END),
      COUNT(*)),1) AS severe_rate
FROM speed_binned
GROUP BY speed_bin
ORDER BY speed_bin;

""" # end _sql
from google.colab.sql import bigquery as _bqsqlcell
df = _bqsqlcell.run(_sql)
df
```

**Answer:** When grouping speed limits into broader ranges, low-speed roads show the lowest likelihood of severe injuries, while mid-speed zones exhibit a modest but noticeable increase in severe outcomes. High-speed areas appear safer only because far fewer crashes occur there. Overall, the pattern reinforces that even small increases in operating speed can raise the probability of severe injury, even when absolute rates remain low.

## **4.4 Road Risk Severity Index (RRSI)</b></font>**

The Road Risk Severity Index (RRSI) is a proposed tool offered to outline areas that are considered to be major crash zones. The RRSI calculates a risk score by assigning weights to specific roads based on crash frequency, injury severity, and time-of-day. Each of these components is first normalized to a common 0 to 100 scale to ensure fair comparison across all locations. Crash frequency is given the greatest weight at 40%, followed by average injury severity at 30%, maximum injury severity at 20%, and time-of-day risk at 10%. These weights reflect both how often crashes occur and the level of harm and risk associated with them.


Index Breakdown:

1. Crash Frequency — How many crashes happen at each location.

2. Injury Severity — How serious those crashes are, using a numeric scale for injury levels.

3. Time-of-Day Risk — Whether crashes happen during high-risk hours (nighttime or rush hour).

4. Normalization & Weighting — scale each component to the same 0–100 range and combine them using weights that reflect their importance.



```python
# sql_engine: bigquery
# output_variable: df
# start _sql
_sql = """
WITH AccidentFrequency AS (
  SELECT
    road_name,
    cross_street_name,
    COUNT(DISTINCT report_number) AS total_accidents
  FROM `ba775-fall25-b04.crash_reporting_dataset.merged_data`
  WHERE road_name IS NOT NULL
    AND cross_street_name IS NOT NULL
  GROUP BY
    road_name,
    cross_street_name
),


SeverityScores AS (
  SELECT
    report_number,
    road_name,
    cross_street_name,
    CASE
      WHEN LOWER(driver_injury_severity) = 'fatal injury' THEN 5
      WHEN LOWER(driver_injury_severity) = 'suspected serious injury' THEN 4
      WHEN LOWER(driver_injury_severity) = 'suspected minor injury' THEN 3
      WHEN LOWER(driver_injury_severity) = 'possible injury' THEN 2
      WHEN LOWER(driver_injury_severity) = 'no apparent injury' THEN 1
      ELSE 0
    END AS driver_score,
    CASE
      WHEN LOWER(nm_injury_severity) = 'fatal injury' THEN 5
      WHEN LOWER(nm_injury_severity) = 'suspected serious injury' THEN 4
      WHEN LOWER(nm_injury_severity) = 'suspected minor injury' THEN 3
      WHEN LOWER(nm_injury_severity) = 'possible injury' THEN 2
      WHEN LOWER(nm_injury_severity) = 'no apparent injury' THEN 1
      ELSE 0
    END AS nm_score
  FROM `ba775-fall25-b04.crash_reporting_dataset.merged_data`
  WHERE road_name IS NOT NULL
    AND cross_street_name IS NOT NULL
),

CombinedSeverityPerCrash AS (
  SELECT
    report_number,
    road_name,
    cross_street_name,
    CASE
      WHEN driver_score > 0 AND nm_score > 0 THEN (driver_score + nm_score) / 2.0
      WHEN driver_score > 0 THEN CAST(driver_score AS FLOAT64)
      WHEN nm_score > 0 THEN CAST(nm_score AS FLOAT64)
      ELSE 0.0
    END AS combined_severity_score
  FROM SeverityScores
),

InjurySeverity AS (
  SELECT
    road_name,
    cross_street_name,
    AVG(combined_severity_score) AS average_injury_severity_score,
    MAX(combined_severity_score) AS max_injury_severity_score
  FROM CombinedSeverityPerCrash
  GROUP BY
    road_name,
    cross_street_name
),

TimeOfDayRisk AS (
  SELECT
    road_name,
    cross_street_name,
    EXTRACT(HOUR FROM crash_datetime) AS crash_hour,
    COUNT(DISTINCT report_number) AS crashes_in_hour
  FROM `ba775-fall25-b04.crash_reporting_dataset.merged_data`
  WHERE crash_datetime IS NOT NULL
    AND road_name IS NOT NULL
    AND cross_street_name IS NOT NULL
  GROUP BY
    road_name,
    cross_street_name,
    crash_hour
),

TimeOfDayAggregated AS (
  SELECT
    road_name,
    cross_street_name,
    SAFE_DIVIDE(
      SUM(
        CASE
          WHEN crash_hour IN (18, 19, 20, 21, 22, 23, 0, 1) THEN crashes_in_hour
          WHEN crash_hour IN (7, 8, 9, 17, 18) THEN crashes_in_hour * 0.7
          ELSE crashes_in_hour * 0.3
        END
      ),
      NULLIF(SUM(crashes_in_hour), 0)
    ) AS high_risk_hour_ratio,
    SUM(crashes_in_hour) AS total_crashes_tod
  FROM TimeOfDayRisk
  GROUP BY
    road_name,
    cross_street_name
),


CombinedMetrics AS (
  SELECT
    COALESCE(af.road_name, isv.road_name, tod.road_name) AS road_name,
    COALESCE(af.cross_street_name, isv.cross_street_name, tod.cross_street_name) AS cross_street_name,
    COALESCE(af.total_accidents, 0) AS total_accidents,
    COALESCE(isv.average_injury_severity_score, 0) AS avg_severity_score,
    COALESCE(isv.max_injury_severity_score, 0) AS max_severity_score,
    COALESCE(tod.high_risk_hour_ratio, 0) AS high_risk_hour_ratio,
    COALESCE(tod.total_crashes_tod, 0) AS total_crashes_tod
  FROM AccidentFrequency af
  FULL OUTER JOIN InjurySeverity isv
    ON af.road_name = isv.road_name
    AND af.cross_street_name = isv.cross_street_name
  FULL OUTER JOIN TimeOfDayAggregated tod
    ON COALESCE(af.road_name, isv.road_name) = tod.road_name
    AND COALESCE(af.cross_street_name, isv.cross_street_name) = tod.cross_street_name
),


NormalizationFactors AS (
  SELECT
    MAX(total_accidents) AS max_accidents,
    MIN(total_accidents) AS min_accidents,
    MAX(avg_severity_score) AS max_avg_severity,
    MIN(avg_severity_score) AS min_avg_severity,
    MAX(max_severity_score) AS max_max_severity,
    MIN(max_severity_score) AS min_max_severity,
    MAX(high_risk_hour_ratio) AS max_risk_ratio,
    MIN(high_risk_hour_ratio) AS min_risk_ratio
  FROM CombinedMetrics
),


RRSICalculation AS (
  SELECT
    cm.road_name,
    cm.cross_street_name,
    cm.total_accidents,
    cm.avg_severity_score,
    cm.max_severity_score,
    cm.high_risk_hour_ratio,


    CASE
      WHEN nf.max_accidents = nf.min_accidents THEN 50.0
      ELSE SAFE_DIVIDE(cm.total_accidents - nf.min_accidents,
                       NULLIF(nf.max_accidents - nf.min_accidents, 0)) * 100.0
    END AS normalized_frequency,


    CASE
      WHEN nf.max_avg_severity = nf.min_avg_severity THEN 50.0
      ELSE SAFE_DIVIDE(cm.avg_severity_score - nf.min_avg_severity,
                       NULLIF(nf.max_avg_severity - nf.min_avg_severity, 0)) * 100.0
    END AS normalized_avg_severity,


    CASE
      WHEN nf.max_max_severity = nf.min_max_severity THEN 50.0
      ELSE SAFE_DIVIDE(cm.max_severity_score - nf.min_max_severity,
                       NULLIF(nf.max_max_severity - nf.min_max_severity, 0)) * 100.0
    END AS normalized_max_severity,


    CASE
      WHEN nf.max_risk_ratio = nf.min_risk_ratio THEN 50.0
      ELSE SAFE_DIVIDE(cm.high_risk_hour_ratio - nf.min_risk_ratio,
                       NULLIF(nf.max_risk_ratio - nf.min_risk_ratio, 0)) * 100.0
    END AS normalized_time_risk
  FROM CombinedMetrics cm
  CROSS JOIN NormalizationFactors nf
),

FinalRRSI AS (
  SELECT
    road_name,
    cross_street_name,
    total_accidents,
    ROUND(avg_severity_score, 2) AS avg_severity_score,
    ROUND(max_severity_score, 2) AS max_severity_score,
    ROUND(high_risk_hour_ratio, 3) AS high_risk_hour_ratio,
    ROUND(normalized_frequency, 2) AS normalized_frequency,
    ROUND(normalized_avg_severity, 2) AS normalized_avg_severity,
    ROUND(normalized_max_severity, 2) AS normalized_max_severity,
    ROUND(normalized_time_risk, 2) AS normalized_time_risk,

    (
      normalized_frequency * 0.40 +
      normalized_avg_severity * 0.30 +
      normalized_max_severity * 0.20 +
      normalized_time_risk * 0.10
    ) AS rrsi_raw
  FROM RRSICalculation
)


SELECT
  road_name,
  cross_street_name,
  total_accidents,
  ROUND(avg_severity_score, 2) AS avg_severity_score,
  ROUND(max_severity_score, 2) AS max_severity_score,
  ROUND(high_risk_hour_ratio, 3) AS high_risk_hour_ratio,
  ROUND(normalized_frequency, 2) AS normalized_frequency,
  ROUND(normalized_avg_severity, 2) AS normalized_avg_severity,
  ROUND(normalized_max_severity, 2) AS normalized_max_severity,
  ROUND(normalized_time_risk, 2) AS normalized_time_risk,
  ROUND(rrsi_raw, 2) AS rrsi_score,
  CASE
    WHEN rrsi_raw >= 75 THEN 'CRITICAL'
    WHEN rrsi_raw >= 50 THEN 'HIGH'
    WHEN rrsi_raw >= 25 THEN 'MODERATE'
    ELSE 'LOW'
  END AS risk_category
FROM FinalRRSI
ORDER BY
  rrsi_score DESC,
  total_accidents DESC;

""" # end _sql
from google.colab.sql import bigquery as _bqsqlcell
df = _bqsqlcell.run(_sql)
df
```

**Answer:** The RRSI results highlight intersections where crash frequency, injury severity, and high-risk timing come together to create elevated danger. Locations along major corridors like Georgia Ave, New Hampshire Ave, and Rockville Pike repeatedly appear at the top of the ranking, showing that these high-volume arterials experience not only more crashes but also more severe injury profiles and a greater share of incidents during late-night or rush-hour periods. Some locations with only a few crashes still score high due to very severe injuries or because all crashes occurred during risky hours, illustrating why a composite index is more informative than crash counts alone. Overall, the RRSI helps identify intersections where both crash patterns and crash consequences suggest a meaningful safety concern, making them strong candidates for targeted engineering or enforcement interventions.
